```python
# Import necessary libraries
import pandas as pd
import numpy as np

# Define a list of subject IDs
subjects = ['04', '05', '06', '07', '08', '09', '11', '12', '13']

# Read the CSV file into a DataFrame using the first subject ID from the list and the specified file path
df = pd.read_csv("/Users/yillysaurus/Downloads/data/sub-" + subjects[0] + "/func2/events.tsv", delimiter="\t")

# Fill any missing values in the DataFrame with 0
df = df.fillna(0)

# Import the 'math' module for mathematical operations
import math

# Assign the 'onset' values to the 'Timepoint' column
df = df.assign(Timepoint=df['onset'].values)

# For each row in the DataFrame, check if the 'trial_type' is 'inter_trials',
# and if so, set the corresponding 'Timepoint' value to 0
for index, row in df.iterrows():
    if row['trial_type'] == 'inter_trials':
        df.at[index, 'Timepoint'] = 0

# Round the 'Timepoint' values to the nearest integer, then divide them by 2, and round again to the nearest integer
df.Timepoint = [math.ceil(x) for x in df.Timepoint]
df.Timepoint = [x / 2 for x in df.Timepoint]
df.Timepoint = [math.ceil(x) for x in df.Timepoint]

# Find the 'Timepoint' values corresponding to the rows where 'num_loss_cards' is equal to 5
# and store them as a list of integers in 'high_cards'
high_cards = df[df['num_loss_cards'] == 5].Timepoint.values.tolist()
high_cards = sorted(set(high_cards))
high_list = [int(x) for x in high_cards]

# Convert the 'high_list' to a space-separated string and print the result
highly_list = ' '.join(str(item) for item in high_list)
print('Max_card_index=', highly_list)

# Find the 'Timepoint' values corresponding to the rows where 'num_loss_cards' is equal to 1
# and store them as a list of integers in 'low_cards'
low_cards = df[df['num_loss_cards'] == 1].Timepoint.values.tolist()
low_cards = sorted(set(low_cards))
low_list = [int(x) for x in low_cards]

# Convert the 'low_list' to a space-separated string and print the result
lowly_list = ' '.join(str(item) for item in low_list)
print('Min_card_index=', lowly_list)

# Print the 'Timepoint' value corresponding to the row where 'gain_amount' is equal to the maximum value in the column
print('Highest_win_index=', df[df['gain_amount'] == max(df['gain_amount'])].Timepoint.values)

# Find the 'Timepoint' values corresponding to the rows where 'gain_amount' is equal to 1
# and store them as a list of integers in 'lowest_win'
lowest_win = df[df['gain_amount'] == 1].Timepoint.values.tolist()
lowest_win = sorted(set(lowest_win))
lowest_list = [int(x) for x in lowest_win]

# Convert the 'lowest_list' to a space-separated string and print the result
lowestly_list = ' '.join(str(item) for item in lowest_list)
print('Lowest_win_index=', lowestly_list)

# Find the 'Timepoint' values corresponding to the first and final indices of non-zero 'Timepoint' values
first_indices = df[(df['Timepoint'] != 0) & (df['Timepoint'].shift(-1) != 0) & (df['Timepoint'] != 210) & (df['Timepoint'].shift(1) == 0)].index
final_indices = df[(df['Timepoint'] != 0) & (df['Timepoint'].shift(1) != 0) & (df['Timepoint'] != 210) & (df['Timepoint'].shift(-1) == 0)].index

# Find the 'Timepoint' values corresponding to the fail indices (non-zero followed by zero 'Timepoint' values)
fail_indices = df[(df['Timepoint'] != 0) & (df['Timepoint'].shift(-1) == 0) & (df['Timepoint'] != 210) & (df['Timepoint'].shift(1) == 0)].index

# Print the 'Timepoint' value corresponding to the second-to-last row in the DataFrame
print('Reward_index=', df[-2:-1].Timepoint.values)

# Print the 'Timepoint' values corresponding to rows where 'trial_type' is 'rest_block'
print('Rest_index=', df[df['trial_type'] == 'rest_block'].Timepoint.values)

# Extract the 'Timepoint' values corresponding to the first and final indices, and convert them to space-separated strings
list_try_first = df['Timepoint'][first_indices].tolist()
first_list = [int(x) for x in list_try_first]
firstly_list = ' '.join(str(item) for item in first_list)

list_try_final = df['Timepoint'][final_indices].tolist()
final_list = [int(x) for x in list_try_final]
finally_list = ' '.join(str(item) for item in final_list)

# Print the first and final lists
print('First list=', firstly_list)
print('Final list=', finally_list)

# Extract the 'Timepoint' values corresponding to the fail indices and convert them to a space-separated string
list_try_fail = df['Timepoint'][fail_indices].tolist()
fail_list = [int(x) for x in list_try_fail]
failly_list = ' '.join(str(item) for item in fail_list)

# Print the fail list
print('Fail list=', failly_list)

```
